```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from scipy.integrate import trapz
import random as rn
```

```python
#%% FORMATO GRRAFICAS
def formato_grafico(ax=None, decimal=None):
    # Configuración personalizada de la gráfica
    if ax is None:
        ax = plt.gca()

    #fuente
    fuente = 'serif'
    ax.xaxis.label.set_size(14)
    ax.yaxis.label.set_size(14)
    ax.tick_params(axis='both', labelsize=12)
    ax.title.set_size(14)
    ax.title.set_family(fuente)
    ax.yaxis.label.set_family(fuente)
    ax.xaxis.label.set_family(fuente)

   # Cambiar configuración de fuentes matemáticas
    plt.rcParams['mathtext.fontset'] = 'cm'  # Cambia a 'cm', 'stixsans', 'dejavusans', etc.


    # Locator y sublocadores
    ax.xaxis.set_major_locator(ticker.AutoLocator())
    ax.xaxis.set_minor_locator(ticker.AutoMinorLocator())
    ax.yaxis.set_major_locator(ticker.AutoLocator())
    ax.yaxis.set_minor_locator(ticker.AutoMinorLocator())

    # Configurar los ticks
    ax.tick_params(axis='x', direction='in', which='both', bottom=True, top=True)
    ax.tick_params(axis='y', direction='in', which='both', left=True, right=True)
    ax.tick_params(axis='x', which='minor', direction='in', bottom=True, top=True)
    ax.tick_params(axis='y', which='minor', direction='in', left=True, right=True)


    # Establecer el formato del número en el eje y a 0 en lugar de 0.0
    if decimal == 0:
        ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:.0f}'.format(x) if x != 0 else '0'))
        ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:.1f}'.format(x) if x != 0 else '0'))
    elif decimal == 1:
        ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:.0f}'.format(x) if x != 0 else '0'))
        ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:.0f}'.format(x) if x != 0 else '0'))
```

```python
#%% PROBABILIDAD
f_max = 1
t = np.linspace(6, 18, 13)

f = np.where(t <= 14, f_max / (14 - 6) * (t - 6),
             f_max - (f_max / (18 - 14)) * (t - 14))


# Continuo
integral_f = np.trapz(f, t)
f_normalized = f / integral_f


# Usando valor medio
t_mid = t[:-1] + 0.5
f = np.where(t <= 14, f_max / (14 - 6) * (t - 6),
             f_max - (f_max / (18 - 14)) * (t - 14))
f_mid = np.interp(t_mid, t, f)
area_original = np.sum(f_mid)
f_mid_normalized = f_mid / area_original
t_steps = np.arange(6, 18)
f_step = np.repeat(f_mid_normalized, 2)
t_step_edges = np.repeat(t_steps, 2)[1:-1]
t_step_edges = np.insert(t_step_edges, 0, 6)
t_step_edges = np.append(t_step_edges, 18)

#Usando integral
t_fino = np.linspace(6, 18, 10000)
f_fino = np.where(t_fino <= 14, f_max / (14 - 6) * (t_fino - 6),
                  f_max - (f_max / (18 - 14)) * (t_fino - 14)
                  )
prob_integral = []
intervalos = 12
p = len(t_fino) // intervalos
for i in range(12):
    t_interval = t_fino[i*p:(i+1)*p]
    f_interval = f_fino[i*p:(i+1)*p]
    integral_value = trapz(f_interval, t_interval)
    prob_integral.append(integral_value)
prob_integral = np.array(prob_integral)
prob_integral /= np.sum(prob_integral)


# Graficar
plt.figure(figsize=(8, 5))
bar_plot = plt.bar(t[:-1], prob_integral, width=1, align='edge', color='green', alpha=0.3, label='Probabilidad discretizada usando integral')
step_plot = plt.step(t_step_edges, f_step, color='b', where='post', label="Probabilidad discretizada usando valor medio")
line_plot = plt.plot(t, f_normalized, 'r-', label="Probabilidad continua")
plt.title('Probabilidad de Lluvia en Bogotá')
plt.xlabel('Hora del día ' + r'$t$' + ' (hora)')
plt.ylabel('Probabilidad de lluvia ' + r'$f\ (t)$')
plt.grid(True)
plt.xticks(np.arange(6, 19, 1))
plt.legend(handles=[line_plot[0], step_plot[0], bar_plot], labels=["Probabilidad continua", "Probabilidad discretizada usando valor medio", "Probabilidad discretizada usando integral"], framealpha=0)
plt.tight_layout()
formato_grafico()
plt.show()
```

```python
#%% PROBABILIDAD ACUMULADA
f_cumulative_normalized = np.cumsum(f_normalized) * (t[1] - t[0])
f_cumulative_step = np.cumsum(f_mid_normalized) * (t_mid[1] - t_mid[0])
f_step_cumulative = np.repeat(f_cumulative_step, 2)

# Graficar
plt.figure(figsize=(8, 5))
plt.step(t_step_edges, f_step_cumulative, color='b', where='post', label="Probabilidad acumulada discretizada")
plt.plot(t, f_cumulative_normalized, 'r-', label="Probabilidad acumulada continua")
plt.title('Probabilidad Acumulada de Lluvia en Bogotá')
plt.xlabel('Hora del día ' + r'$t\ (hora)$')
plt.ylabel('Probabilidad acumulada de lluvia ' + r'$F(t)$')
plt.grid(True)
plt.xticks(np.arange(6, 19, 1))
plt.legend(framealpha=0.3)
plt.tight_layout()
plt.show()
```

```python
#%% EXPERIMENTO VARIOS DIAS
rn.seed(121)
pdf = np.insert(f_step[1::2], 0, 0)
dim_pdf = len(pdf)
F = np.zeros(dim_pdf)


for i in range(1, dim_pdf):
    F[i] = F[i-1] + pdf[i]

bins = np.linspace(6, 18, 13)

fig, axs = plt.subplots(2, 2, figsize=(15, 16))

N_values = [30, 365, 1000, 10**6]

for idx, N in enumerate(N_values):
    histograma = np.zeros(13)

    for i in range(N):
        r = rn.random()
        for j in range(1, dim_pdf):
            if r >= F[j-1] and r < F[j]:
                bingo = j - 1
                break
        histograma[bingo] += 1

    ax = axs[idx // 2, idx % 2]

    ax.step(bins, histograma, where='post', color='k')

    ax.set_xlabel('Hora del día ' + r'$t\ (hora)$')
    ax.set_ylabel('Frecuencia')
    ax.set_title(f'Histograma para {N} dias')
    ax.grid(True)
    formato_grafico(ax)
    ax.set_xticks(np.arange(6, 19, 1))

plt.suptitle('Histograma de lluvia a diferentes horas para N dias', fontsize=16, font='serif')
plt.show()
```

```python
#%% COMPARACION SIMULACION Y MODELO

rn.seed(121)
pdf = np.insert(f_step[1::2], 0, 0)
dim_pdf = len(pdf)
F = np.zeros(dim_pdf)


for i in range(1, dim_pdf):
    F[i] = F[i-1] + pdf[i]

bins = np.linspace(6, 18, 13)

fig, axs = plt.subplots(2, 2, figsize=(15, 16))

N_values = [30, 365, 1000, 10**6]

for idx, N in enumerate(N_values):
    histograma = np.zeros(13)

    for i in range(N):
        r = rn.random()
        for j in range(1, dim_pdf):
            if r >= F[j-1] and r < F[j]:
                bingo = j - 1
                break
        histograma[bingo] += 1

    ax = axs[idx // 2, idx % 2]

    histograma_normalizado = histograma
    f_step_normalizado =  f_step*N

    ax.step(t_step_edges, f_step_normalizado, color='r', where='post', label="Prediccion del modelo")
    ax.step(bins, histograma_normalizado, where='post', color='k', label = "Datos simulacion")

    ax.legend(loc='upper left', framealpha=0.3)
    ax.set_xlabel('Hora del dia ' + r'$t\ (hora)$')
    ax.set_ylabel('Frecuencia')
    ax.set_title(f'Histograma para {N} dias')
    ax.grid(True)
    formato_grafico(ax)
    ax.set_xticks(np.arange(6, 19, 1))

plt.suptitle('Histograma de lluvia a diferentes horas para N dias y comparacion con el modelo', fontsize=16, font='serif')
plt.show()
```

```python
#%% PROBABILIDAD DE NO LLOVER

rn.seed(121)
p = 0.7

pdf = np.insert(f_step[1::2], 0, 0)
dim_pdf = len(pdf)
F = np.zeros(dim_pdf)

for i in range(1, dim_pdf):
    F[i] = F[i-1] + pdf[i]

bins = np.linspace(6, 18, 13)

fig, axs = plt.subplots(1, 2, figsize=(15, 6))

N_values = [30, 100]

for idx, N in enumerate(N_values):
    histograma = np.zeros(13)
    d_n=0
    for i in range(N):
        # Determinamos si llueve ese día con una probabilidad del 70%
        if rn.random() < 0.7:  # 70% de probabilidad de que llueva ese día
            # Si llueve, generamos un valor aleatorio para la distribución horaria
            r = rn.random()
            for j in range(1, dim_pdf):
                if r >= F[j-1] and r < F[j]:
                    bingo = j - 1
                    break
            histograma[bingo] += 1

        else:
            d_n += 1

    ax = axs[idx]
    ax.step(bins, histograma, where='post', color='k', label="Datos simulación")
    ax.text(0.35, 0.93, f'Si llovió en {N-d_n} días', transform=ax.transAxes, ha='left', va='top', fontsize=14, font='serif')
    ax.text(0.35, 0.89, f'No llovió en {d_n} días', transform=ax.transAxes, ha='left', va='top', fontsize=14, font='serif')
    ax.set_ylim(top=int(max(histograma)*1.4))
    ax.set_xlabel('Hora del día ' + r'$t\ (hora)$')
    ax.set_ylabel('Frecuencia')
    ax.set_title(f'Histograma para {N} dias')
    ax.grid(True)
    formato_grafico(ax)
    ax.set_xticks(np.arange(6, 19, 1))

plt.suptitle('Histograma de lluvia a diferentes horas para N dias con probabilidad de llover del 70% por dia', fontsize=16, font='serif')
plt.show()
```

```python
# Configuración inicial
rn.seed(121)
p = 0.7

pdf = np.insert(f_step[1::2], 0, 0)
dim_pdf = len(pdf)
F = np.zeros(dim_pdf)

# Construcción de la distribución acumulada
for i in range(1, dim_pdf):
    F[i] = F[i-1] + pdf[i]

# Definición de los límites del histograma
bins = np.linspace(6, 18, 13)

# Configuración de la figura y subplots
fig, axs = plt.subplots(1, 2, figsize=(15, 6))

# Número de días de simulación
N = 10**6
histograma = np.zeros(13)
d_n = 0  # Contador de días sin lluvia

# Simulación de los días de lluvia
for i in range(N):
    # Determinamos si llueve ese día con una probabilidad del 70%
    if rn.random() < p:  # 70% de probabilidad de que llueva ese día
        # Si llueve, generamos un valor aleatorio para la distribución horaria
        r = rn.random()
        for j in range(1, dim_pdf):
            if r >= F[j-1] and r < F[j]:
                bingo = j - 1
                break
        histograma[bingo] += 1
    else:
        # Incrementamos el contador de días sin lluvia
        d_n += 1

# Histograma normalizado por el total de días simulados N
histograma_normalizado_N = histograma / N

# Histograma normalizado por los días en los que realmente llovió (N - d_n)
histograma_normalizado_lluvia = histograma / (N - d_n)

# Graficado de ambos histogramas en subplots
for idx, (ax, histograma_normalizado, label) in enumerate(zip(
    axs,
    [histograma_normalizado_N, histograma_normalizado_lluvia],
    ["Normalizado con N", "Normalizado con N - d_n"]
)):
    ax.step(bins, histograma_normalizado, where='post', color='k', label="Datos simulación")
    ax.step(t_step_edges, f_step, color='r', where='post', label="Predicción del modelo")

    # Texto explicativo
    ax.text(0.05, 0.89, f'Si llovió en {N - d_n} días', transform=ax.transAxes, ha='left', va='top', fontsize=11, font='serif')
    ax.text(0.05, 0.86, f'No llovió en {d_n} días', transform=ax.transAxes, ha='left', va='top', fontsize=11, font='serif')

    # Personalización del gráfico
    ax.legend(loc='upper left', framealpha=0.3)
    ax.set_xlabel('Hora del día ' + r'$t\ (hora)$')
    ax.set_ylabel('Frecuencia')
    ax.set_title(f'Histograma  {label}')
    ax.grid(True)
    formato_grafico(ax)
    ax.set_xticks(np.arange(6, 19, 1))

# Título principal de la figura
plt.suptitle('Histograma normalizado de lluvia a diferentes horas para N días con probabilidad de llover del 70% por día', fontsize=16, font='serif')
plt.show()
```
