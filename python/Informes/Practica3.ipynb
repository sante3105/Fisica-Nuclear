```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.ticker as ticker
from matplotlib import font_manager
import numpy as np
from scipy.optimize import curve_fit
import os
from datetime import datetime

#%% FORMATO GRAFICAS
def formato_grafico(ax=None, decimalx=None, decimaly=None, legend_loc='upper right'):
    if ax is None:
        ax = plt.gca()

    fuente = 'sans'
    ax.title.set_size(14)
    ax.xaxis.label.set_size(16)
    ax.yaxis.label.set_size(16)
    ax.title.set_family(fuente)
    ax.yaxis.label.set_family(fuente)
    ax.xaxis.label.set_family(fuente)
    ax.tick_params(axis='both', labelsize=14)

    plt.rcParams['mathtext.fontset'] = 'cm'  
    plt.rcParams['mathtext.rm'] = 'sans'  
    plt.rcParams['mathtext.default'] = 'rm'

    ax.xaxis.set_major_locator(ticker.AutoLocator())
    ax.xaxis.set_minor_locator(ticker.AutoMinorLocator())
    ax.yaxis.set_major_locator(ticker.AutoLocator())
    ax.yaxis.set_minor_locator(ticker.AutoMinorLocator())

    ax.tick_params(axis='x', direction='in', which='both', bottom=True, top=True)
    ax.tick_params(axis='y', direction='in', which='both', left=True, right=True)
    ax.tick_params(axis='x', which='minor', direction='in', bottom=True, top=True)
    ax.tick_params(axis='y', which='minor', direction='in', left=True, right=True)

    # Configurar la notacin cientfica
    if decimalx is not None:
        ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{x:.{decimalx}f}' if x != 0 else '0'))

    if decimaly is not None:
        ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{x:.{decimaly}f}' if x != 0 else '0'))

    if ax.get_legend() is not None:
        ax.legend(fontsize=16)

    font = font_manager.FontProperties(family='sans', style='normal', size=16)

    handles, labels = ax.get_legend_handles_labels()
    if handles and labels:
        ax.legend(loc=legend_loc, handles=handles, labels=labels, edgecolor='1', framealpha=0.5, prop=font)

#%% Funcion para el ajuste gaussiano
def gauss(x, A, B, C, mu, sigma):
    return A + B * (x - mu) + C * np.exp(-(x - mu) ** 2 / (2 * sigma ** 2))

def gauss_ajuste(x, y):
    media = sum(x * y) / sum(y)
    sigma = np.sqrt(sum(y * (x - media) ** 2) / sum(y))
    popt, pcov = curve_fit(gauss, x, y, p0=[1, 1, min(y), media, sigma])
    return popt, pcov

def Ajuste_gaussiano(canal, cuentas, extremo_inferior, extremo_superior, a_0=0, a_1=1, unidades='canal', titulo='Ajuste Gaussiano', etiqueta = 'Ajuste pico', ylim=None):
    x = canal[extremo_inferior:extremo_superior]
    y = cuentas[extremo_inferior:extremo_superior]

    x = x * a_1 + a_0

    popt, pcov = gauss_ajuste(x, y)
    A, B, C, mu, sigma = popt

    x_ajuste = np.linspace(extremo_inferior, extremo_superior, 100) * a_1 + a_0
    y_ajuste = A + B * (x_ajuste - mu) + C * np.exp(-1 / 2 * ((x_ajuste - mu) / sigma) ** 2)
    recta_ajuste = A + B * (x_ajuste - mu)

    FWHM = 2.35482 * sigma
    uncertainties = np.sqrt(np.diag(pcov))

    print("\nPico:     ", titulo)
    print("b0:        {:.0f}  {:.0f}".format(A, uncertainties[0]))
    print("b1:        {:.3f}  {:.3f}".format(B, uncertainties[1]))
    print("M:         {:.0f}  {:.0f}".format(C, uncertainties[2]))
    print(":         {:.2f}  {:.2f}".format(mu, uncertainties[3]))
    print(":         {:.2f}  {:.2f}".format(abs(sigma), abs(uncertainties[4])))
    print("FWHM:      {:.2f}  {:.2f}".format(abs(FWHM), abs(FWHM * (uncertainties[4] / sigma))))
    sigma = sigma
    uncertainties[4] = uncertainties[4]
    I = abs((2 * np.pi) ** (1 / 2) * C * sigma)
    dI = I*np.sqrt((uncertainties[2]/C)**2 + (uncertainties[4]/sigma)**2)
    print("Intensidad:      {:.0f}  {:.0f}".format(abs(I), abs(dI)))


    plt.figure(figsize=(6, 5))

    plt.step(x, y, 'k', label='{}'.format(etiqueta), where='mid')
    plt.plot(x_ajuste, y_ajuste, '--r', linewidth=2,label=r'$b_0 + b_1 (x - \mu) + M \exp \left[-\frac{(x-\mu)^2}{2\sigma^2} \right]$')
    plt.plot(x_ajuste, recta_ajuste, '--b', linewidth=2, label=r'$b_0 + b_1 (x - \mu) $')
    plt.grid()
    plt.title("")
    plt.ylabel(r'$I_{\gamma}$' + ' [cuentas / canal]')
    if unidades == 'canal':
        plt.xlabel(r'canal')
    elif unidades == 'keV':
        plt.xlabel(r'$E_{\gamma}$' + ' (keV)')
        
    if ylim:
        plt.ylim(ylim)
    
    formato_grafico()
    plt.tight_layout()
    plt.ticklabel_format(axis='y', style='sci', scilimits=(0,0))
    # Guardar la imagen
    save_path = r"Informe3/Imagenes"
    filename = f"Practica3_{titulo.replace(' ', '_')}.png"
    full_path = os.path.join(save_path, filename)
    plt.savefig(full_path, dpi=300, bbox_inches='tight')
    

    plt.show()


    
#%% Funcion doble ajuste gaussiano
def double_gauss(x, A, B, C1, mu1, sigma1, C2, mu2, sigma2):
    return (A + B * (x - mu1) + 
            C1 * np.exp(-((x - mu1) ** 2) / (2 * sigma1 ** 2)) + 
            C2 * np.exp(-((x - mu2) ** 2) / (2 * sigma2 ** 2)))

# Ajuste de la doble gaussiana
def double_gauss_ajuste(x, y):
    media = sum(x * y) / sum(y)
    sigma = np.sqrt(sum(y * (x - media) ** 2) / sum(y))

    # Estimaciones iniciales: dos picos cercanos en la distribucin
    p0 = [1, 1, max(y), media - sigma, sigma, max(y) / 2, media + sigma, sigma]

    popt, pcov = curve_fit(double_gauss, x, y, p0=p0)
    return popt, pcov

# Funcin para realizar el ajuste y graficar
def Ajuste_doble_gaussiano(canal, cuentas, extremo_inferior, extremo_superior, 
                           a_0=0, a_1=1, unidades='canal', titulo='Ajuste Doble Gaussiano', etiqueta = 'etiqueta', ylim=None):
    x = canal[extremo_inferior:extremo_superior]
    y = cuentas[extremo_inferior:extremo_superior]
    x = x * a_1 + a_0

    popt, pcov = double_gauss_ajuste(x, y)
    A, B, C1, mu1, sigma1, C2, mu2, sigma2 = popt

    x_ajuste = np.linspace(extremo_inferior, extremo_superior, 100) * a_1 + a_0
    y_ajuste = double_gauss(x_ajuste, *popt)
    recta_ajuste = A + B * (x_ajuste - mu1)

    FWHM1 = 2.35482 * sigma1
    FWHM2 = 2.35482 * sigma2
    uncertainties = np.sqrt(np.diag(pcov))

    
    # Impresin de resultados
    print("\nPico:     ", titulo)
    print("b0:        {:.0f}  {:.0f}".format(A, uncertainties[0]))
    print("b1:        {:.3f}  {:.3f}".format(B, uncertainties[1]))
    print("M1:        {:.0f}  {:.0f}".format(C1, uncertainties[2]))
    print("1:        {:.2f}  {:.2f}".format(mu1, uncertainties[3]))
    print("1:        {:.2f}  {:.2f}".format(abs(sigma1), abs(uncertainties[4])))
    print("FWHM1:     {:.2f}  {:.2f}".format(abs(FWHM1), abs(FWHM1 * (uncertainties[4] / sigma1))))
    print("M2:        {:.0f}  {:.0f}".format(C2, uncertainties[5]))
    print("2:        {:.2f}  {:.2f}".format(mu2, uncertainties[6]))
    print("2:        {:.2f}  {:.2f}".format(abs(sigma2), abs(uncertainties[7])))
    print("FWHM2:     {:.2f}  {:.2f}".format(abs(FWHM2), abs(FWHM2 * (uncertainties[7] / sigma2))))

    sigma1 = sigma1
    uncertainties[4] = uncertainties[4]
    I1 = abs((2 * np.pi) ** (1 / 2) * C1 * sigma1)
    dI1 = I1*np.sqrt((uncertainties[2]/C1)**2 + (uncertainties[4]/sigma1)**2)

    sigma2 = sigma2
    uncertainties[7] = uncertainties[7]
    I2 = abs((2 * np.pi) ** (1 / 2) * C2 * sigma2)
    dI2 = I2*np.sqrt((uncertainties[5]/C2)**2 + (uncertainties[7]/sigma2)**2)
    
    print("Intensidad1:      {:.0f}  {:.0f}".format(abs(I1), abs(dI1)))
    print("Intensidad2:      {:.0f}  {:.0f}".format(abs(I2), abs(dI2)))
    print("IntensidadT:      {:.0f}  {:.0f}".format(abs(I1+I2), abs(dI1+dI2)))
    # Graficar
    plt.figure(figsize=(5, 5))
    plt.step(x, y, 'k', label='{}'.format(etiqueta), where='mid')
    plt.plot(x_ajuste, y_ajuste, '--r', linewidth=2, 
             label=r'$b_0 + b_1 (x - \mu) + M_1 \exp \left[-\frac{(x-\mu_1)^2}{2\sigma_1^2} \right]+ M_2 \exp \left[-\frac{(x-\mu_2)^2}{2\sigma_2^2} \right]$')
    plt.plot(x_ajuste, recta_ajuste, '--b', linewidth=2, label=r'$b_0 + b_1 (x - \mu_1) $')
    plt.grid()
    plt.ylabel(r'$I_{\gamma}$' + ' [cuentas / canal]')
    
    if unidades == 'canal':
        plt.xlabel(r'canal')
    elif unidades == 'keV':
        plt.xlabel(r'$E_{\gamma}$' + ' [keV]')
    
    if ylim:
        plt.ylim(ylim)

    formato_grafico()
    plt.tight_layout()
    
    # Guardar la imagen
    save_path = r"Informe3/Imagenes"
    filename = f"Practica3_{titulo.replace(' ', '_')}.png"
    full_path = os.path.join(save_path, filename)
    plt.savefig(full_path, dpi=300, bbox_inches='tight')
    

    plt.show()
    
#%% Rutas de los archivos

# Funcin para cargar los datos
def cargar_datos(ruta):
    return pd.read_csv(ruta, sep=r'\s+', names=["canal", "cuentas"])

# Rutas de los archivos
ruta_Eu     = r"Informe3/Data/calibracionEu-Ge8K.xy"
ruta_NaCoCs = r"Informe3/Data/calibracionNaCoCoCs-Ge8K.xy"
# Cargar los datos
datos_Eu = cargar_datos(ruta_Eu)
datos_NaCoCs = cargar_datos(ruta_NaCoCs)



#%%
plt.figure(figsize=(10, 5))

# Graficar los datos

# Posiciones de los picos y etiquetas
picos = [411, 1897, 3307, 3577, 3757, 6983]
etiquetas = ['122.1 keV', '661.7 keV', '1173.2 keV', '1274.5 keV','1332.5 keV', '2505.7 keV']

# Graficar lneas punteadas y textos
for pico, etiqueta in zip(picos, etiquetas):
    plt.axvline(x=pico, linestyle='dashed', color='k', alpha=0.9)
    plt.text(pico + 50, 13**4, etiqueta, rotation=90)
    
    
plt.step(datos_NaCoCs["canal"], datos_NaCoCs["cuentas"], color='r', label=r'$^{22}\text{Na} + ^{60}\text{Co} + ^{57}\text{Co} + ^{137}\text{Cs}$', where='mid')

plt.xlabel(r'Canal')
plt.ylabel(r'$I_{\gamma}$' + ' [cuentas/canal]')
plt.legend()
plt.tight_layout()
plt.grid()
plt.ylim([1, 1000000])
formato_grafico()
plt.yscale('log')


# Guardar la imagen
save_path = r"Informe3/Imagenes"
filename = "Practica3_NaCoCoCs.png"
full_path = os.path.join(save_path, filename)
plt.savefig(full_path, dpi=300, bbox_inches='tight')

plt.show()

#%%
Ajuste_gaussiano(datos_NaCoCs["canal"], datos_NaCoCs["cuentas"], extremo_inferior=395, extremo_superior=420, unidades='canal', titulo=r'122.1 keV 57Co', etiqueta=r'Pico 122.1 keV $^{57}Co$', ylim = [0,500000])
Ajuste_gaussiano(datos_NaCoCs["canal"], datos_NaCoCs["cuentas"], extremo_inferior=1875, extremo_superior=1915, unidades='canal', titulo=r'661.7 keV 137Cs', etiqueta=r'Pico 661.7 keV $^{137}Cs$', ylim = [0,180000])
Ajuste_gaussiano(datos_NaCoCs["canal"], datos_NaCoCs["cuentas"], extremo_inferior=3290, extremo_superior=3330, unidades='canal', titulo=r'1173.2 60Co', etiqueta=r'Pico 1173.2 keV $^{60}Co$', ylim = [0,26000])
Ajuste_gaussiano(datos_NaCoCs["canal"], datos_NaCoCs["cuentas"], extremo_inferior=3570, extremo_superior=3605, unidades='canal', titulo=r'1274.5 22Na', etiqueta=r'Pico 1274.5 keV $^{22}Na$', ylim = [0,4000])
Ajuste_gaussiano(datos_NaCoCs["canal"], datos_NaCoCs["cuentas"], extremo_inferior=3730, extremo_superior=3765, unidades='canal', titulo=r'1332.5 60Co', etiqueta=r'Pico 1332.5 keV $^{60}Co$', ylim = [0,25000])
Ajuste_gaussiano(datos_NaCoCs["canal"], datos_NaCoCs["cuentas"], extremo_inferior=6965, extremo_superior=7005, unidades='canal', titulo=r'2505.7 suma', etiqueta=r'Pico 2505.7 keV suma', ylim = [0,800])


#%% Crear la grfica
plt.figure(figsize=(10, 5))

# Posiciones de los picos y etiquetas
picos = [314, 364, 1020, 2063, 2726, 3581, 4467]
etiquetas = ['86.5 keV', '105.3 keV', '344.3 keV', '723.3 keV', '964 keV', '1274.5 keV', '1596.5 keV']

# Graficar lneas punteadas y textos
for i, (pico, etiqueta) in enumerate(zip(picos, etiquetas)):
    plt.axvline(x=pico, linestyle='dashed', color='k', alpha=0.9)
    
    if i == 0:  # Para la primera etiqueta, colocarla a la izquierda
        plt.text(pico - 20, 46**3, etiqueta, rotation=90, ha='right')
    elif i == 1:  # Para la primera etiqueta, colocarla a la izquierda
        plt.text(pico + 80, 46**3, etiqueta, rotation=90)        
    else:
        plt.text(pico + 60, 46**3, etiqueta, rotation=90)
        
# Graficar los datos
plt.step(datos_Eu["canal"], datos_Eu["cuentas"], color='blue', label=r'$^{152,154,155}\text{Eu}$', where='mid')

#plt.xlim([0, 500])
plt.xlabel(r'Canal')
plt.ylabel(r'$I_{\gamma}$' + ' [cuentas/canal]')
plt.legend()
plt.tight_layout()
plt.grid()
plt.ylim([1,1000000])
formato_grafico()
plt.yscale('log')
# Guardar la imagen
save_path = r"Informe3/Imagenes"
filename = "Practica3_Eu.png"
full_path = os.path.join(save_path, filename)
plt.savefig(full_path, dpi=300, bbox_inches='tight')

plt.show()


#%%
Ajuste_gaussiano(datos_Eu["canal"], datos_Eu["cuentas"], extremo_inferior=300, extremo_superior=330, unidades='canal', titulo=r'86.5 keV 155Eu', etiqueta=r'Pico 86.5 keV $^{155}Eu$', ylim = [0,28000])
Ajuste_gaussiano(datos_Eu["canal"], datos_Eu["cuentas"], extremo_inferior=350, extremo_superior=380, unidades='canal', titulo=r'105.3 keV 154Eu', etiqueta=r'Pico 105.3 keV $^{154}Eu$', ylim = [0,28000])
Ajuste_gaussiano(datos_Eu["canal"], datos_Eu["cuentas"], extremo_inferior=1005, extremo_superior=1030, unidades='canal', titulo=r'344.3 keV 152Eu', etiqueta=r'Pico 344.3 keV $^{152}Eu$', ylim = [0,28000])
Ajuste_gaussiano(datos_Eu["canal"], datos_Eu["cuentas"], extremo_inferior=2050, extremo_superior=2080, unidades='canal', titulo=r'723.3 keV 154Eu', etiqueta=r'Pico 723.3 keV $^{154}Eu$', ylim = [0,6000])
Ajuste_gaussiano(datos_Eu["canal"], datos_Eu["cuentas"], extremo_inferior=2710, extremo_superior=2740, unidades='canal', titulo=r'964.1 keV 152Eu', etiqueta=r'Pico 964.1 keV $^{152}Eu$', ylim = [0,5000])
Ajuste_gaussiano(datos_Eu["canal"], datos_Eu["cuentas"], extremo_inferior=3565, extremo_superior=3595, unidades='canal', titulo=r'1274.5 keV 154Eu', etiqueta=r'Pico 1274.5 keV $^{154}Eu$', ylim = [0,6000])
Ajuste_gaussiano(datos_Eu["canal"], datos_Eu["cuentas"], extremo_inferior=4455, extremo_superior=4485, unidades='canal', titulo=r'1596.5 keV 154Eu', etiqueta=r'Pico 1596.5 keV $^{154}Eu$', ylim = [0,500])  



#%%

# Energa en keV
E_gamma_keV = [122.1, 661.7, 1173.2, 1274.5, 1332.5, 2505.7, 86.5, 105.3, 344.3, 723.3, 964.1, 1274.5, 1596.5]

# Mu y su error
E_gamma_canal = [410.61, 1896.49, 3307.23, 3586.65, 3746.46, 6982.47, 312.16, 363.68, 1020.10, 2062.91, 2725.83, 3580.55, 4467.27]
dE_gamma_canal = [0.03, 0.02, 0.02, 0.03, 0.02, 0.05, 0.02, 0.01, 0.01, 0.01, 0.02, 0.01, 0.03]


# Definir funcin lineal
def linear(x, a, b):
    return a + b * x

# Ajuste por mnimos cuadrados ponderado
popt, pcov = curve_fit(linear, E_gamma_canal, E_gamma_keV, sigma=dE_gamma_canal, absolute_sigma=True)
a, b = popt
uncertainties = np.sqrt(np.diag(pcov))  # Incertezas en los parmetros a y b

# Generar datos para la recta ajustada
x_fit = np.linspace(min(E_gamma_canal), max(E_gamma_canal), 100)
y_fit = linear(x_fit, *popt)

# Mostrar resultados
print("\nAjuste lineal (Canal vs. Energa en keV):")
print("a0:        {:.2f}  {:.2f}".format(a, uncertainties[0]))
print("a1:        {:.4f}  {:.2f}".format(b, uncertainties[1]))

# Graficar
fig, ax = plt.subplots(figsize=(10, 5))

ax.errorbar(E_gamma_canal, E_gamma_keV, xerr=dE_gamma_canal, fmt='o', label="Datos (ajustes de picos)", color='k', capsize=3)
ax.plot(x_fit, y_fit, '-r', label=r"$E_{\gamma} \ [keV] = a_0 + a_1 \times canal$")

ax.set_ylabel(r'$E_{\gamma}$ [keV]')
ax.set_xlabel('canal')
ax.grid()

# Formato grfico
formato_grafico(legend_loc='upper left')
plt.tight_layout()

# Guardar la imagen
save_path = r"Informe3/Imagenes"
filename = "Practica3_Calibracion.png"
full_path = os.path.join(save_path, filename)
plt.savefig(full_path, dpi=300, bbox_inches='tight')

plt.show()



# Sigma y su error
#sigma = [1.64, 1.45, 2.20, 2.51, 2.54, 2.60, 4.08, 3.24, 1.36, 1.41, 1.74, 2.06, 2.25, 2.45, 2.70]
#error_sigma = [0.04, 0.02, 0.02, 0.02, 0.03, 0.02, 0.06, 0.06, 0.02, 0.01, 0.01, 0.01, 0.02, 0.01, 0.04]

# Intensidad y su error
intensidad = [1130387, 857362, 627840, 103917, 14804, 4072, 3828, 56612, 48092, 60116, 16022, 13747, 21471, 1749]
error_intensidad = [31033, 29381, 7343, 1213, 229, 76, 83, 886, 665, 488, 138, 151, 169, 31]


#%%
# Datos experimentales
E_gamma_keV = np.array([122.1, 661.7, 1173.2, 1274.5, 1332.5, 2505.7, 86.5, 105.3, 344.3, 723.3, 964.1, 1596.5])
fwhm = np.array([3.87, 5.17, 5.92, 5.98, 6.11, 7.63, 3.19, 3.32, 4.10, 4.84, 5.29, 6.37])
dfwhm = np.array([0.08, 0.05, 0.05, 0.07, 0.05, 0.13, 0.04, 0.04, 0.03, 0.03, 0.05, 0.09])

# Conversin a unidades de keV
fwhm_kev = 0.3637 * fwhm
dfwhm_kev = 0.3637 * dfwhm  # Sin la constante -27.11

# Definir modelos de ajuste
def fwhm_sqrt_model(E_gamma, f0, f1):
    return f0 + f1 * np.sqrt(E_gamma)

def fwhm_linear_model(E_gamma, l0, l1):
    return l0 + l1 * E_gamma

# Ajuste con sqrt(E_gamma)
popt_sqrt, pcov_sqrt = curve_fit(fwhm_sqrt_model, E_gamma_keV, fwhm_kev, sigma=dfwhm_kev, absolute_sigma=True)
f0, f1 = popt_sqrt
uncert_sqrt = np.sqrt(np.diag(pcov_sqrt))

# Ajuste con E_gamma lineal
popt_linear, pcov_linear = curve_fit(fwhm_linear_model, E_gamma_keV, fwhm_kev, sigma=dfwhm_kev, absolute_sigma=True)
l0, l1 = popt_linear
uncert_linear = np.sqrt(np.diag(pcov_linear))

# Generar datos para la curva ajustada
E_gamma_fit = np.linspace(min(E_gamma_keV), max(E_gamma_keV), 100)
fwhm_sqrt_fit = fwhm_sqrt_model(E_gamma_fit, f0, f1)
fwhm_linear_fit = fwhm_linear_model(E_gamma_fit, l0, l1)

# Clculo del coeficiente de correlacin
fwhm_sqrt_ajuste = fwhm_sqrt_model(E_gamma_keV, f0, f1)
fwhm_linear_ajuste = fwhm_linear_model(E_gamma_keV, l0, l1)

r_sqrt = np.corrcoef(fwhm_kev, fwhm_sqrt_ajuste)[0, 1]
r_linear = np.corrcoef(fwhm_kev, fwhm_linear_ajuste)[0, 1]

# Graficar
fig, ax = plt.subplots(figsize=(10, 5))

ax.errorbar(E_gamma_keV, fwhm_kev, yerr=dfwhm_kev, fmt='o', label="Datos experimentales", color='k', capsize=3)
ax.plot(E_gamma_fit, fwhm_sqrt_fit, '-r', label=rf"$\text{{FWHM}} = f_0 + f_1 \sqrt{{E_\gamma}},  R={r_sqrt:.3f}$")
ax.plot(E_gamma_fit, fwhm_linear_fit, '--b', label=rf"$\text{{FWHM}} = l_0 + l_1 E_\gamma,  R={r_linear:.3f}$")

ax.set_xlabel(r'$E_{\gamma}$ [keV]')
ax.set_ylabel(r'FWHM [keV]')
ax.grid()

# Formato grfico
formato_grafico(legend_loc='upper left')
plt.tight_layout()

# Guardar la imagen
save_path = r"Informe3/Imagenes"
filename_fwhm = "Practica3_FWHM_Ajuste.png"
full_path_fwhm = os.path.join(save_path, filename_fwhm)
plt.savefig(full_path_fwhm, dpi=300, bbox_inches='tight')

plt.show()



# Mostrar resultados del ajuste
print("\nAjuste de FWHM en funcin de E:")
print(f"f0: {f0:.3f}  {uncert_sqrt[0]:.3f}")
print(f"f1: {f1:.3f}  {uncert_sqrt[1]:.3f}")
print(f"Coeficiente de correlacin R: {r_sqrt:.4f}")

print("\nAjuste de FWHM en funcin de E:")
print(f"l0: {l0:.3f}  {uncert_linear[0]:.3f}")
print(f"l1: {l1:.3f}  {uncert_linear[1]:.3f}")
print(f"Coeficiente de correlacin R: {r_linear:.4f}")



#%%
#                [155, 154, 152, 154, 152, 154, 154]
E_gamma_keV =      [86.5, 105.3, 344.3, 723.3, 964.1, 1274.5,1596.5]
T_1div2 = np.array([4.76,  8.59,  13.6,  8.59,  13.6,   8.59,  8.59]) # aos
T_1div2 = [i*365 for i in T_1div2] #dias
tau = T_1div2 / np.log(2)
t_0 = [13072006]*7
A_0 =     np.array([12.8,  8.45,  7.65,  8.45,  7.65,   8.45,  8.45])

t = 29052018  # Fecha de referencia
b_gamma = np.array([3.1,    2.2, 26.57,  20.2, 14.63,  35.0,  1.81])
d_b_gamma = np.array([0.001, 0.001, 0.001, 0.001, 0.001, 0.001, 0.001])

I = np.array([56612, 48092, 60116, 16022, 13747, 21471, 1749])
d_I = np.array([886, 665, 488, 138, 151, 169, 31])
dt = [1, 1, 1, 1, 1, 1, 1]
# Convertir fechas a objetos datetime.date
fechas_0 = [datetime.strptime(str(fecha), "%d%m%Y").date() for fecha in t_0]
fecha_t = datetime.strptime(str(t), "%d%m%Y").date()

# Calcular diferencias en das
d_t = np.array([abs((fecha_t - fecha).days) for fecha in fechas_0])

# Calcular A
A = A_0 * np.exp(-d_t / tau)

# Calcular A_gamma y su incertidumbre
A_gamma = A * b_gamma/100
d_A_gamma = A_gamma * np.sqrt((d_b_gamma / b_gamma) ** 2)  # Asumiendo solo error de b_gamma

# Calcular eficiencia y su incertidumbre
epsilon = I / (A_gamma*dt)
d_epsilon = epsilon * np.sqrt((d_I / I) ** 2 + (d_A_gamma / A_gamma) ** 2)


print(f"{'E_gamma (keV)':>12} | {'A (kBq)':>15} | {'b_gamma':>10} | {'A_gamma (kBq)':>15} | {'I':>10} | {'':>15}")
print("-" * 120)
for i in range(len(E_gamma_keV)):
    print(f"{E_gamma_keV[i]:>12.1f} | "
          f"{A[i]:>10.2f} | "
          f"{b_gamma[i]:.2f}({d_b_gamma[i]:.3f}) | "
          f"{A_gamma[i]:>15.2f}({d_A_gamma[i]:.2f}) | "
          f"{I[i]:>10}({d_I[i]:.0f}) | "
          f"{epsilon[i]:>10.5f}({d_epsilon[i]:.3f})")

epsilon_norm = epsilon / np.max(epsilon)
d_epsilon_norm = d_epsilon / np.max(epsilon)

plt.figure(figsize=(10, 5))
plt.errorbar(E_gamma_keV, epsilon_norm, yerr=d_epsilon_norm,c='k', fmt='o', capsize=5)
plt.xlabel(r'$E_{\gamma}$ [keV]')
plt.ylabel('Eficiencia Normalizada')
plt.grid(True, linestyle='--', alpha=0.6)
plt.yscale('log')

formato_grafico(decimaly=1)
# Guardar la imagen
save_path = r"Informe3/Imagenes"
filename = "Practica3_Eficiencia.png"
full_path = os.path.join(save_path, filename)
plt.savefig(full_path, dpi=300, bbox_inches='tight')

plt.show()

#%%

def cargar_datos(ruta):
    return pd.read_csv(ruta, sep=r'\s+', names=["canal", "cuentas"])

ruta_NaCoCs = r"Informe3/Data/calibracionNaCoCoCs-Ge8K.xy"
datos_NaCoCs = cargar_datos(ruta_NaCoCs)

def cargar_datos(ruta):
    return pd.read_csv(ruta, sep=r'\s+', skiprows=1, names=["canal", "cuentas"])

# Rutas de los archivos
ruta_22Na600  = r"Informe2/Data/NaI_22Na_600s.dat"
ruta_57Co300  = r"Informe2/Data/NaI_57Co_300s.dat"
ruta_60Co600  = r"Informe2/Data/NaI_60Co_600s.dat"
ruta_137Cs600 = r"Informe2/Data/NaI_137Cs_600s.dat"
ruta_Fondo300 = r"Informe2/Data/NaI_Fondo_300s.dat"
ruta_Fondo600 = r"Informe2/Data/NaI_Fondo_600s.dat"



datos_22Na = cargar_datos(ruta_22Na600)
datos_57Co = cargar_datos(ruta_57Co300)
datos_60Co = cargar_datos(ruta_60Co600)
datos_137Cs = cargar_datos(ruta_137Cs600)
datos_Fondo300 = cargar_datos(ruta_Fondo300)
datos_Fondo600 = cargar_datos(ruta_Fondo600)

#####################

datos_22Na["cuentas"] = datos_22Na["cuentas"] - datos_Fondo600["cuentas"]
datos_57Co["cuentas"] = datos_57Co["cuentas"] - datos_Fondo300["cuentas"]
datos_60Co["cuentas"] = datos_60Co["cuentas"] - datos_Fondo600["cuentas"]
datos_137Cs["cuentas"] = datos_137Cs["cuentas"] - datos_Fondo600["cuentas"]



#####################

a =-59.61
b = 3.1077

datos_NaCoCs["canal"] = [-26.39 + 0.3631*i for i in datos_NaCoCs["canal"]]
datos_57Co["canal"] = [a + b*i for i in datos_57Co["canal"]]
datos_60Co["canal"] = [a + b*i for i in datos_60Co["canal"]]
datos_22Na["canal"] = [a + b*i for i in datos_22Na["canal"]]
datos_137Cs["canal"] = [-a + b*i for i in datos_137Cs["canal"]]

datos_NaCoCs["cuentas"] = [i/max(datos_NaCoCs["cuentas"]) for i in datos_NaCoCs["cuentas"]]
datos_57Co["cuentas"] = [i/max(datos_57Co["cuentas"]) for i in datos_57Co["cuentas"]]
datos_60Co["cuentas"] = [i/max(datos_60Co["cuentas"]) for i in datos_60Co["cuentas"]]
datos_22Na["cuentas"] = [i/max(datos_22Na["cuentas"]) for i in datos_22Na["cuentas"]]
datos_137Cs["cuentas"] = [i/max(datos_137Cs["cuentas"]) for i in datos_137Cs["cuentas"]]


plt.figure(figsize=(10, 8))

plt.step(datos_NaCoCs["canal"], datos_NaCoCs["cuentas"], color='r', label=r'$^{22}\text{Na} + ^{60}\text{Co} + ^{57}\text{Co} + ^{137}\text{Cs}$, detector Ge', where='mid')
plt.step(datos_57Co["canal"], datos_57Co["cuentas"], color='k', label=r'$^{57}\text{Co}$, detector NaI', where='mid')


plt.xlabel(r'$E_{\gamma}$ [keV]')
plt.ylabel(r'$I_{\gamma}$' + ' [normalizada]')
plt.legend()
plt.tight_layout()
plt.grid()
plt.xlim([40, 220])
plt.ylim([0, 1.2])
formato_grafico()

# Guardar la imagen
save_path = r"Informe3/Imagenes"
filename = "Practica3_Comparacion_1.png"
full_path = os.path.join(save_path, filename)
plt.savefig(full_path, dpi=300, bbox_inches='tight')

plt.show()

#%%

def cargar_datos(ruta):
    return pd.read_csv(ruta, sep=r'\s+', names=["canal", "cuentas"])

ruta_NaCoCs = r"Informe3/Data/calibracionNaCoCoCs-Ge8K.xy"
datos_NaCoCs = cargar_datos(ruta_NaCoCs)

def cargar_datos(ruta):
    return pd.read_csv(ruta, sep=r'\s+', skiprows=1, names=["canal", "cuentas"])

# Rutas de los archivos
ruta_22Na600  = r"Informe2/Data/NaI_22Na_600s.dat"
ruta_57Co300  = r"Informe2/Data/NaI_57Co_300s.dat"
ruta_60Co600  = r"Informe2/Data/NaI_60Co_600s.dat"
ruta_137Cs600 = r"Informe2/Data/NaI_137Cs_600s.dat"
ruta_Fondo300 = r"Informe2/Data/NaI_Fondo_300s.dat"
ruta_Fondo600 = r"Informe2/Data/NaI_Fondo_600s.dat"


datos_22Na = cargar_datos(ruta_22Na600)
datos_57Co = cargar_datos(ruta_57Co300)
datos_60Co = cargar_datos(ruta_60Co600)
datos_137Cs = cargar_datos(ruta_137Cs600)
datos_Fondo300 = cargar_datos(ruta_Fondo300)
datos_Fondo600 = cargar_datos(ruta_Fondo600)

#####################

datos_22Na["cuentas"] = datos_22Na["cuentas"] - datos_Fondo600["cuentas"]
datos_57Co["cuentas"] = datos_57Co["cuentas"] - datos_Fondo300["cuentas"]
datos_60Co["cuentas"] = datos_60Co["cuentas"] - datos_Fondo600["cuentas"]
datos_137Cs["cuentas"] = datos_137Cs["cuentas"] - datos_Fondo600["cuentas"]



#####################

datos_NaCoCs["canal"] = [-26.39 + 0.3631*i for i in datos_NaCoCs["canal"]]
datos_60Co["canal"] = [-22 + 3.1177*i for i in datos_60Co["canal"]]
datos_22Na["canal"] = [-21 + 3.1177*i for i in datos_22Na["canal"]]

# Definir constantes
E1 = 1173  # Energa en keV
E2 = 1333  # Energa en keV
m_e = 511  # Masa del electrn en keV

# Funciones para calcular el borde Compton y la retrodispersin
def calcular_borde_compton(E):
    return 2 * E**2 / (2 * E + m_e)

# Calcular bordes Compton
E_C1 = calcular_borde_compton(E1)
E_C2 = calcular_borde_compton(E2)

# Graficar espectros
plt.figure(figsize=(10, 8))

plt.step(datos_NaCoCs["canal"], datos_NaCoCs["cuentas"], color='k', label=r'$^{22}\text{Na} + ^{60}\text{Co} + ^{57}\text{Co} + ^{137}\text{Cs}$, detector Ge', where='mid')
plt.step(datos_60Co["canal"], datos_60Co["cuentas"], color='b', label=r'$^{60}\text{Co}$, detector NaI', where='mid')
plt.step(datos_22Na["canal"], datos_22Na["cuentas"], color='r', label=r'$^{22}\text{Na}$, detector NaI', where='mid')

# Aadir lneas del borde Compton
plt.axvline(E_C1, color='g', linestyle='--', label=f'Borde Compton {E1} keV')
plt.axvline(E_C2, color='m', linestyle='--', label=f'Borde Compton {E2} keV')


# Etiquetas y formato
plt.xlabel(r'$E_{\gamma}$ [keV]')
plt.ylabel(r'$I_{\gamma}$')
plt.legend()
plt.tight_layout()
plt.grid()
plt.xlim([800, 1600])
plt.ylim([1, 2000])
formato_grafico(legend_loc='upper left')

# Guardar la imagen
save_path = r"Informe3/Imagenes"
filename = "Practica3_Comparacion_2.png"
full_path = os.path.join(save_path, filename)
plt.savefig(full_path, dpi=300, bbox_inches='tight')

plt.show()

#%%
```
