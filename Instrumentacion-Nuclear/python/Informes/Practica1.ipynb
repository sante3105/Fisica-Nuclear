```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.ticker as ticker
from matplotlib import font_manager
import numpy as np
from scipy.optimize import curve_fit
import os

#%% FORMATO GRAFICAS
def formato_grafico(ax=None, decimalx=None, decimaly=None, legend_loc='upper right'):
    if ax is None:
        ax = plt.gca()

    fuente = 'sans'
    ax.title.set_size(14)
    ax.xaxis.label.set_size(16)
    ax.yaxis.label.set_size(16)
    ax.title.set_family(fuente)
    ax.yaxis.label.set_family(fuente)
    ax.xaxis.label.set_family(fuente)
    ax.tick_params(axis='both', labelsize=14)

    plt.rcParams['mathtext.fontset'] = 'cm'  
    plt.rcParams['mathtext.rm'] = 'sans'  
    plt.rcParams['mathtext.default'] = 'rm'

    ax.xaxis.set_major_locator(ticker.AutoLocator())
    ax.xaxis.set_minor_locator(ticker.AutoMinorLocator())
    ax.yaxis.set_major_locator(ticker.AutoLocator())
    ax.yaxis.set_minor_locator(ticker.AutoMinorLocator())

    ax.tick_params(axis='x', direction='in', which='both', bottom=True, top=True)
    ax.tick_params(axis='y', direction='in', which='both', left=True, right=True)
    ax.tick_params(axis='x', which='minor', direction='in', bottom=True, top=True)
    ax.tick_params(axis='y', which='minor', direction='in', left=True, right=True)

    # Configurar la notacin cientfica
    if decimalx is not None:
        ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{x:.{decimalx}f}' if x != 0 else '0'))
    if decimaly is not None:
        ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{x:.{decimaly}f}' if x != 0 else '0'))

    if ax.get_legend() is not None:
        ax.legend(fontsize=16)

    font = font_manager.FontProperties(family='sans', style='normal', size=16)
    handles, labels = ax.get_legend_handles_labels()
    if handles and labels:
        ax.legend(loc=legend_loc, handles=handles, labels=labels, edgecolor='1', framealpha=0.5, prop=font)

#%% Funciones para ajustes

# Ajuste Gaussiano
def gauss(x, A, B, C, mu, sigma):
    return A + B * (x - mu) + C * np.exp(-(x - mu) ** 2 / (2 * sigma ** 2))

def gauss_ajuste(x, y):
    media = sum(x * y) / sum(y)
    sigma = np.sqrt(sum(y * (x - media) ** 2) / sum(y))
    popt, pcov = curve_fit(gauss, x, y, p0=[1, 1, min(y), media, sigma])
    return popt, pcov

def Ajuste_gaussiano(canal, cuentas, extremo_inferior, extremo_superior, 
                     a_0=0, a_1=1, unidades='canal', titulo='Ajuste Gaussiano', ylim=None):
    x = canal[extremo_inferior:extremo_superior]
    y = cuentas[extremo_inferior:extremo_superior]

    x = x * a_1 + a_0

    popt, pcov = gauss_ajuste(x, y)
    A, B, C, mu, sigma = popt

    x_ajuste = np.linspace(extremo_inferior, extremo_superior, 100) * a_1 + a_0
    y_ajuste = A + B * (x_ajuste - mu) + C * np.exp(-(x_ajuste - mu) ** 2 / (2 * sigma ** 2))
    recta_ajuste = A + B * (x_ajuste - mu)

    FWHM = 2.35482 * sigma
    uncertainties = np.sqrt(np.diag(pcov))

    print("\nPico:     ", titulo)
    print("b0:        {:.0f}  {:.0f}".format(A, uncertainties[0]))
    print("b1:        {:.3f}  {:.3f}".format(B, uncertainties[1]))
    print("M:         {:.0f}  {:.0f}".format(C, uncertainties[2]))
    print(":         {:.2f}  {:.2f}".format(mu, uncertainties[3]))
    print(":         {:.2f}  {:.2f}".format(abs(sigma), abs(uncertainties[4])))
    print("FWHM:      {:.2f}  {:.2f}".format(abs(FWHM), abs(FWHM * (uncertainties[4] / sigma))))

    I = abs((2 * np.pi) ** (1/2) * C * sigma)
    dI = I * np.sqrt((uncertainties[2]/C)**2 + (uncertainties[4]/sigma)**2)
    print("Intensidad:      {:.0f}  {:.0f}".format(abs(I), abs(dI)))

    plt.figure(figsize=(10, 5))
    plt.step(x, y, 'k', label='{}'.format(titulo), where='mid')
    plt.plot(x_ajuste, y_ajuste, '--r', linewidth=2, 
             label=r'$b_0 + b_1 (x - \mu) + M \exp \left[-\frac{(x-\mu)^2}{2\sigma^2} \right]$')
    plt.plot(x_ajuste, recta_ajuste, '--b', linewidth=2, 
             label=r'$b_0 + b_1 (x - \mu) $')
    plt.grid()
    plt.title("")
    plt.ylabel(r'$I_{\gamma}$ [cuentas / canal]')
    if unidades == 'canal':
        plt.xlabel(r'canal')
    elif unidades == 'keV':
        plt.xlabel(r'$E_{\gamma}$ (keV)')
    if ylim:
        plt.ylim(ylim)

    formato_grafico()
    plt.tight_layout()

    # Guardar la imagen
    save_path = r"Informe1/Imagenes"
    filename = f"Practica1_{titulo.replace(' ', '_')}.png"
    full_path = os.path.join(save_path, filename)
    plt.savefig(full_path, dpi=300, bbox_inches='tight')
    plt.show()

# Ajuste doble Gaussiano
def double_gauss(x, A, B, C1, mu1, sigma1, C2, mu2, sigma2):
    return (A + B * (x - mu1) +
            C1 * np.exp(-((x - mu1) ** 2) / (2 * sigma1 ** 2)) +
            C2 * np.exp(-((x - mu2) ** 2) / (2 * sigma2 ** 2)))

def double_gauss_ajuste(x, y):
    media = sum(x * y) / sum(y)
    sigma = np.sqrt(sum(y * (x - media) ** 2) / sum(y))
    p0 = [1, 1, max(y), media - sigma, sigma, max(y)/2, media + sigma, sigma]
    popt, pcov = curve_fit(double_gauss, x, y, p0=p0)
    return popt, pcov

def Ajuste_doble_gaussiano(canal, cuentas, extremo_inferior, extremo_superior, 
                           a_0=0, a_1=1, unidades='canal', titulo='Ajuste Doble Gaussiano',
                           ylim=None):
    x = canal[extremo_inferior:extremo_superior]
    y = cuentas[extremo_inferior:extremo_superior]
    x = x * a_1 + a_0

    popt, pcov = double_gauss_ajuste(x, y)
    A, B, C1, mu1, sigma1, C2, mu2, sigma2 = popt

    x_ajuste = np.linspace(extremo_inferior, extremo_superior, 100) * a_1 + a_0
    y_ajuste = double_gauss(x_ajuste, *popt)
    recta_ajuste = A + B * (x_ajuste - mu1)

    FWHM1 = 2.35482 * sigma1
    FWHM2 = 2.35482 * sigma2
    uncertainties = np.sqrt(np.diag(pcov))

    print("\nPico:     ", titulo)
    print("b0:        {:.0f}  {:.0f}".format(A, uncertainties[0]))
    print("b1:        {:.3f}  {:.3f}".format(B, uncertainties[1]))
    print("M1:        {:.0f}  {:.0f}".format(C1, uncertainties[2]))
    print("1:        {:.2f}  {:.2f}".format(mu1, uncertainties[3]))
    print("1:        {:.2f}  {:.2f}".format(abs(sigma1), abs(uncertainties[4])))
    print("FWHM1:     {:.2f}  {:.2f}".format(abs(FWHM1), abs(FWHM1 * (uncertainties[4] / sigma1))))
    print("M2:        {:.0f}  {:.0f}".format(C2, uncertainties[5]))
    print("2:        {:.2f}  {:.2f}".format(mu2, uncertainties[6]))
    print("2:        {:.2f}  {:.2f}".format(abs(sigma2), abs(uncertainties[7])))
    print("FWHM2:     {:.2f}  {:.2f}".format(abs(FWHM2), abs(FWHM2 * (uncertainties[7] / sigma2))))

    I1 = abs((2 * np.pi) ** (1/2) * C1 * sigma1)
    dI1 = I1 * np.sqrt((uncertainties[2]/C1)**2 + (uncertainties[4]/sigma1)**2)
    I2 = abs((2 * np.pi) ** (1/2) * C2 * sigma2)
    dI2 = I2 * np.sqrt((uncertainties[5]/C2)**2 + (uncertainties[7]/sigma2)**2)

    print("Intensidad1:      {:.0f}  {:.0f}".format(abs(I1), abs(dI1)))
    print("Intensidad2:      {:.0f}  {:.0f}".format(abs(I2), abs(dI2)))
    print("IntensidadT:      {:.0f}  {:.0f}".format(abs(I1+I2), abs(dI1+dI2)))

    plt.figure(figsize=(10, 5))
    plt.step(x, y, 'k', label='{}'.format(titulo), where='mid')
    plt.plot(x_ajuste, y_ajuste, '--r', linewidth=2, 
             label=r'$b_0 + b_1 (x - \mu) + M_1 \exp \left[-\frac{(x-\mu_1)^2}{2\sigma_1^2} \right] + M_2 \exp \left[-\frac{(x-\mu_2)^2}{2\sigma_2^2} \right]$')
    plt.plot(x_ajuste, recta_ajuste, '--b', linewidth=2, 
             label=r'$b_0 + b_1 (x - \mu_1) $')
    plt.grid()
    plt.ylabel(r'$I_{\gamma}$ [cuentas / canal]')
    if unidades == 'canal':
        plt.xlabel(r'canal')
    elif unidades == 'keV':
        plt.xlabel(r'$E_{\gamma}$ [keV]')
    if ylim:
        plt.ylim(ylim)

    formato_grafico()
    plt.tight_layout()

    # Guardar la imagen
    save_path = r"Informe1/Imagenes"
    filename = f"Practica1_{titulo.replace(' ', '_')}.png"
    full_path = os.path.join(save_path, filename)
    plt.savefig(full_path, dpi=300, bbox_inches='tight')
    plt.show()


#%% RUTAS DE LOS ARCHIVOS (adaptado para informe1/Data)
ruta_cs137 = r"Informe1/Data/Cs137_5minNaI.csv"
ruta_co60  = r"Informe1/Data/Co60_10minNaI.csv"
ruta_na22  = r"Informe1/Data/Na22_10minNaI.csv"

# Leer los archivos CSV
datos_cs137 = pd.read_csv(ruta_cs137)
datos_co60  = pd.read_csv(ruta_co60)
datos_na22  = pd.read_csv(ruta_na22)

# Acceder a las columnas
canal_cs137   = datos_cs137['canal']
cuentas_cs137 = datos_cs137['cuentas']
canal_co60    = datos_co60['canal']
cuentas_co60  = datos_co60['cuentas']
canal_na22    = datos_na22['canal']
cuentas_na22  = datos_na22['cuentas']

#%% Grfica comparativa
plt.figure(figsize=(10, 5))
plt.step(canal_cs137, cuentas_cs137, 'k', label=r'$^{137}\text{Cs}$', where='mid')
plt.step(canal_co60, cuentas_co60, 'r', label=r'$^{60}\text{Co}$', where='mid')
plt.step(canal_na22, cuentas_na22, 'b', label=r'$^{22}\text{Na}$', where='mid')

plt.xlabel(r'canal')
plt.ylabel(r'$I_{\gamma}$ [cuentas/canal]')
plt.grid()
plt.xlim(0, max(max(canal_cs137), max(canal_co60), max(canal_na22)))
plt.ylim(0, max(max(cuentas_cs137), max(cuentas_co60), max(cuentas_na22)) * 1.1)
formato_grafico()
plt.tight_layout()

save_path = r"Informe1/Imagenes"
filename = "Practica1_Patrones.png"
full_path = os.path.join(save_path, filename)
plt.savefig(full_path, dpi=300, bbox_inches='tight')
plt.show()

#%% Ajuste Picos
Ajuste_doble_gaussiano(canal_cs137, cuentas_cs137,
                       extremo_inferior=59, extremo_superior=200,
                       unidades='canal',
                       titulo='1er Pico Cs-137')

Ajuste_gaussiano(canal_cs137, cuentas_cs137,
                 extremo_inferior=59, extremo_superior=200,
                 unidades='canal',
                 titulo='1er Pico Cs-137')

Ajuste_gaussiano(canal_cs137, cuentas_cs137,
                 extremo_inferior=1400, extremo_superior=2100,
                 unidades='canal',
                 titulo='2do Pico Cs-137')

Ajuste_gaussiano(canal_na22, cuentas_na22,
                 extremo_inferior=1100, extremo_superior=1600,
                 unidades='canal',
                 titulo='1er Pico Na-22')

Ajuste_gaussiano(canal_na22, cuentas_na22,
                 extremo_inferior=2800, extremo_superior=3600,
                 unidades='canal',
                 titulo='2do Pico Na-22')

Ajuste_doble_gaussiano(canal_co60, cuentas_co60,
                       extremo_inferior=2400, extremo_superior=3900,
                       unidades='canal',
                       titulo='Picos Co-60',
                       ylim=[-2, 2000])

#%% Ajuste Canal vs. Energa (calibracin)
E_gamma_keV = np.array([32.8, 511.0, 661.7, 1173.2, 1274.5, 1332.5])
E_gamma_canal = np.array([90.87, 1290.90, 1661.57, 2873.02, 3131.53, 3256.48])
dE_gamma_canal = np.array([0.08, 0.06, 0.06, 0.19, 0.26, 0.24])

def linear(x, a, b):
    return a + b*x

popt, pcov = curve_fit(linear, E_gamma_canal, E_gamma_keV,
                       sigma=dE_gamma_canal, absolute_sigma=True)
a, b = popt
uncertainties = np.sqrt(np.diag(pcov))

x_fit = np.linspace(min(E_gamma_canal), max(E_gamma_canal), 100)
y_fit = linear(x_fit, *popt)

print("\nAjuste lineal (Canal vs. Energa en keV):")
print("a0:        {:.2f}  {:.2f}".format(a, uncertainties[0]))
print("a1:        {:.4f}  {:.4f}".format(b, uncertainties[1]))

fig, ax = plt.subplots(figsize=(10, 5))
ax.errorbar(E_gamma_canal, E_gamma_keV, xerr=dE_gamma_canal,
            fmt='o', label="Datos (ajustes de picos)", color='k', capsize=3)
ax.plot(x_fit, y_fit, '--r', label=r"$E_{\gamma} \ [keV] = a_0 + a_1 \times canal$")
ax.set_ylabel(r'$E_{\gamma}$ [keV]')
ax.set_xlabel('canal')
ax.grid()

formato_grafico(legend_loc='upper left')
plt.tight_layout()

save_path = r"Informe1/Imagenes"
filename = "Practica1_Calibracion.png"
full_path = os.path.join(save_path, filename)
plt.savefig(full_path, dpi=300, bbox_inches='tight')
plt.show()

# Ajuste de energa en funcin del canal
E_gamma_ajuste = linear(E_gamma_canal, *popt)
delta = E_gamma_keV - E_gamma_ajuste
percent_error = (delta / E_gamma_keV) * 100

dE_gamma_ajuste = np.sqrt((uncertainties[0])**2 + (uncertainties[1]*E_gamma_canal)**2)
d_delta = dE_gamma_ajuste
d_percent_error = (d_delta / E_gamma_keV) * 100

print("\nTabla con los resultados:")
print("------------------------------------------------------------")
print("| E (keV)   | Canal     | E ajuste (keV)   |  (keV)   | % /E  % |")
print("------------------------------------------------------------")
for energia, c, err_c, ajuste, d, perc, d_aj, d_del, d_perc in zip(E_gamma_keV,
                                                                  E_gamma_canal,
                                                                  dE_gamma_canal,
                                                                  E_gamma_ajuste,
                                                                  delta,
                                                                  percent_error,
                                                                  dE_gamma_ajuste,
                                                                  d_delta,
                                                                  d_percent_error):
    print("| {:10.1f} | {:10.2f}({:.2f}) | {:10.2f}({:.2f}) | {:10.2f}({:.2f}) | {:10.2f}({:.2f}) |".format(
        energia, c, err_c, ajuste, d_aj, d, d_del, perc, d_perc))
print("------------------------------------------------------------")

#%% Curva calibrada
canal_cs137_energia = [a + b*i for i in canal_cs137]
canal_co60_energia  = [a + b*i for i in canal_co60]
canal_na22_energia  = [a + b*i for i in canal_na22]

plt.figure(figsize=(10, 5))
plt.step(canal_cs137_energia, cuentas_cs137, 'k', label=r'$^{137}\text{Cs}$', where='mid')
plt.step(canal_co60_energia, cuentas_co60, 'r', label=r'$^{60}\text{Co}$', where='mid')
plt.step(canal_na22_energia, cuentas_na22, 'b', label=r'$^{22}\text{Na}$', where='mid')

plt.xlabel(r'$E_{\gamma}$ [keV]')
plt.ylabel(r'$I_{\gamma}$ [cuentas/canal]')
plt.grid()
plt.xlim(0, max(max(canal_cs137_energia),
                max(canal_co60_energia),
                max(canal_na22_energia)))
plt.ylim(0, max(max(cuentas_cs137),
                max(cuentas_co60),
                max(cuentas_na22)) * 1.1)
formato_grafico()
plt.tight_layout()

save_path = r"Informe1/Imagenes"
filename = "Practica1_Patrones_energia.png"
full_path = os.path.join(save_path, filename)
plt.savefig(full_path, dpi=300, bbox_inches='tight')
plt.show()

#%% Compton 60Co (datos "experimentales" de ejemplo)
E1 = 2873.02
E2 = 3256.48
borde_c_1 = 2400
retrodispersion_1 = 570

plt.figure(figsize=(10, 5))
plt.step(canal_co60, cuentas_co60, 'k', label=r'$^{60}Co$', where='mid')
plt.axvline(x=borde_c_1, color='b', linestyle='--', label=r'Borde Compton $E_{C1}$')
plt.axvline(x=retrodispersion_1, color='r', linestyle='-.', label=r'Retrodispersin $E_{R1}$')
plt.text(borde_c_1 + 10, max(cuentas_co60)*0.7, rf'$E_{{C1}}$ = {borde_c_1:.2f} canal',
         color='b', rotation=90, verticalalignment='bottom', fontsize=14)
plt.text(retrodispersion_1 + 50, max(cuentas_co60)*0.7, rf'$E_{{R1}}$ = {retrodispersion_1:.2f} canal',
         color='r', rotation=90, verticalalignment='bottom', fontsize=14)
plt.xlabel('canal')
plt.ylabel(r'$I_{\gamma}$ [cuentas/canal]')
plt.grid()
plt.xlim(0, max(canal_co60))
plt.ylim(0, max(cuentas_co60)*1.5)

formato_grafico(legend_loc=[0.2,0.5])
plt.tight_layout()

save_path = r"Informe1/Imagenes"
filename = "Practica1_Patron_60Co_canal.png"
full_path = os.path.join(save_path, filename)
plt.savefig(full_path, dpi=300, bbox_inches='tight')
plt.show()

m_e = 511
def calcular_borde_compton(E):
    return 2*E**2 / (2*E + m_e)
def calcular_retrodispersion(E, E_C):
    return E - E_C

E1 = 1158.08
E2 = 1314.09
borde_c_1 = calcular_borde_compton(E1)
borde_c_2 = calcular_borde_compton(E2)
retrodispersion_1 = calcular_retrodispersion(E1, borde_c_1)
retrodispersion_2 = calcular_retrodispersion(E2, borde_c_2)

print("\n Datos experimentales")
print(f"Borde Compton para E1 ({E1} keV): {borde_c_1:.2f} keV")
print(f"Borde Compton para E2 ({E2} keV): {borde_c_2:.2f} keV")
print(f"Energa de retrodispersion para E1: {retrodispersion_1:.2f} keV")
print(f"Energa de retrodispersion para E2: {retrodispersion_2:.2f} keV")

plt.figure(figsize=(10, 5))
canal_co60_energia = [a + b*i for i in canal_co60]
plt.step(canal_co60_energia, cuentas_co60, 'k', label=r'$^{60}Co$' + ' Experimentales', where='mid')
plt.axvline(x=borde_c_1, color='b', linestyle='--', label=r'Borde Compton $E_{C1}$')
plt.axvline(x=borde_c_2, color='b', linestyle='--', label=r'Borde Compton $E_{C2}$')
plt.axvline(x=retrodispersion_1, color='r', linestyle='-.', label=r'Retrodispersion $E_{R1}$')
plt.axvline(x=retrodispersion_2, color='r', linestyle='-.', label=r'Retrodispersion $E_{R2}$')
plt.text(borde_c_1 + 10, max(cuentas_co60)*0.95, rf'$E_{{C1}}$ = {borde_c_1:.2f} keV',
         color='b', rotation=90, verticalalignment='bottom', fontsize=14)
plt.text(borde_c_2 + 10, max(cuentas_co60)*0.95, rf'$E_{{C2}}$ = {borde_c_2:.2f} keV',
         color='b', rotation=90, verticalalignment='bottom', fontsize=14)
plt.text(retrodispersion_1 - 50, max(cuentas_co60)*0.7, rf'$E_{{R1}}$ = {retrodispersion_1:.2f} keV',
         color='r', rotation=90, verticalalignment='bottom', fontsize=14)
plt.text(retrodispersion_2 + 10, max(cuentas_co60)*0.7, rf'$E_{{R2}}$ = {retrodispersion_2:.2f} keV',
         color='r', rotation=90, verticalalignment='bottom', fontsize=14)
plt.xlabel(r'$E_{\gamma}$ [keV]')
plt.ylabel(r'$I_{\gamma}$ [cuentas/canal]')
plt.grid()
plt.xlim(0, max(canal_co60_energia))
plt.ylim(0, max(cuentas_co60)*2)

formato_grafico(legend_loc=[0.2,0.5])
plt.tight_layout()

save_path = r"Informe1/Imagenes"
filename = "Practica1_Patron_60Co.png"
full_path = os.path.join(save_path, filename)
plt.savefig(full_path, dpi=300, bbox_inches='tight')
plt.show()

E1 = 1173.2
E2 = 1332.5
borde_c_1 = calcular_borde_compton(E1)
borde_c_2 = calcular_borde_compton(E2)
retrodispersion_1 = calcular_retrodispersion(E1, borde_c_1)
retrodispersion_2 = calcular_retrodispersion(E2, borde_c_2)

print("\n Datos teoricos")
print(f"Borde Compton para E1 ({E1} keV): {borde_c_1:.2f} keV")
print(f"Borde Compton para E2 ({E2} keV): {borde_c_2:.2f} keV")
print(f"Energa de retrodispersion para E1: {retrodispersion_1:.2f} keV")
print(f"Energa de retrodispersion para E2: {retrodispersion_2:.2f} keV")

plt.figure(figsize=(10, 5))
plt.step(canal_co60_energia, cuentas_co60, 'k', label=r'$^{60}Co$' + ' Teoricos', where='mid')
plt.axvline(x=borde_c_1, color='b', linestyle='--', label=r'Borde Compton $E_{C1}$')
plt.axvline(x=borde_c_2, color='b', linestyle='--', label=r'Borde Compton $E_{C2}$')
plt.axvline(x=retrodispersion_1, color='r', linestyle='-.', label=r'Retrodispersion $E_{R1}$')
plt.axvline(x=retrodispersion_2, color='r', linestyle='-.', label=r'Retrodispersion $E_{R2}$')
plt.text(borde_c_1 + 10, max(cuentas_co60)*0.95, rf'$E_{{C1}}$ = {borde_c_1:.2f} keV',
         color='b', rotation=90, verticalalignment='bottom', fontsize=14)
plt.text(borde_c_2 + 10, max(cuentas_co60)*0.95, rf'$E_{{C2}}$ = {borde_c_2:.2f} keV',
         color='b', rotation=90, verticalalignment='bottom', fontsize=14)
plt.text(retrodispersion_1 - 50, max(cuentas_co60)*0.7, rf'$E_{{R1}}$ = {retrodispersion_1:.2f} keV',
         color='r', rotation=90, verticalalignment='bottom', fontsize=14)
plt.text(retrodispersion_2 + 10, max(cuentas_co60)*0.7, rf'$E_{{R2}}$ = {retrodispersion_2:.2f} keV',
         color='r', rotation=90, verticalalignment='bottom', fontsize=14)
plt.xlabel(r'$E_{\gamma}$ [keV]')
plt.ylabel(r'$I_{\gamma}$ [cuentas/canal]')
plt.grid()
plt.xlim(0, max(canal_co60_energia))
plt.ylim(0, max(cuentas_co60)*2)

formato_grafico(legend_loc=[0.2,0.5])
plt.tight_layout()

save_path = r"Informe1/Imagenes"
filename = "Practica1_Patron_60Co_teorico.png"
full_path = os.path.join(save_path, filename)
plt.savefig(full_path, dpi=300, bbox_inches='tight')
plt.show()

#%% Borde Compton y retrodispersin para Cs-137
E1 = 665.23
borde_c_1 = calcular_borde_compton(E1)
retrodispersion_1 = calcular_retrodispersion(E1, borde_c_1)

print("\n Datos teoricos")
print(f"Borde Compton para E1 ({E1} keV): {borde_c_1:.2f} keV")
print(f"Energa de retrodispersion para E1: {retrodispersion_1:.2f} keV")

canal_cs137_energia = [a + b*i for i in canal_cs137]
plt.figure(figsize=(10, 5))
plt.step(canal_cs137_energia, cuentas_cs137, 'k', label=r'$^{137}Cs$' + ' Teoricos', where='mid')
plt.fill_between(canal_cs137_energia[0:1200], cuentas_cs137[0:1200],
                 color='gray', alpha=0.5, step='mid', label='Region Compton')
plt.axvline(x=borde_c_1, color='b', linestyle='--', label=r'Borde Compton $E_{C1}$')
plt.axvline(x=retrodispersion_1, color='r', linestyle='-.', label=r'Retrodispersion $E_{R1}$')

plt.text(borde_c_1 + 10, max(cuentas_cs137)*0.5,
         rf'$E_{{C1}}$ = {borde_c_1:.2f} keV',
         color='b', rotation=90, verticalalignment='bottom', fontsize=14)
plt.text(retrodispersion_1 + 10, max(cuentas_cs137)*0.5,
         rf'$E_{{R1}}$ = {retrodispersion_1:.2f} keV',
         color='r', rotation=90, verticalalignment='bottom', fontsize=14)

plt.xlabel(r'$E_{\gamma}$ [keV]')
plt.ylabel(r'$I_{\gamma}$ [cuentas/canal]')
plt.grid()
plt.xlim(0, max(canal_cs137_energia))
plt.ylim(0, max(cuentas_cs137)*1.1)
formato_grafico(legend_loc=[0.5,0.5])
plt.tight_layout()

save_path = r"Informe1/Imagenes"
filename = "Practica1_Patron_137Cs_energia.png"
full_path = os.path.join(save_path, filename)
plt.savefig(full_path, dpi=300, bbox_inches='tight')
plt.show()

print('Cuentas Compton:', sum(cuentas_cs137[0:1200]),
      '(',
      sum(cuentas_cs137[0:1205]) - sum(cuentas_cs137[0:1200]),
      ')')
```
