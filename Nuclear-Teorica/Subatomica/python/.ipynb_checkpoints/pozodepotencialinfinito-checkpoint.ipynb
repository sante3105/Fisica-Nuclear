```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
h = 1.054e-34  # Reduced Planck's constant (J·s)
m_e = 9.109e-31  # Mass of the electron (kg)
eV_to_J = 1.602e-19  # Conversion factor from eV to Joules
a = 10e-10  # Width of the potential well in meters (10 Å)

# Functions
def wave_function(n, x, a):
    """Compute the wave function for the nth quantum state."""
    return np.sqrt(1 / 3) * np.sin(n * np.pi * x / a)

def probability_density(n, x, a):
    """Compute the probability density for the nth quantum state."""
    psi = wave_function(n, x, a)
    return np.abs(psi)**2

def energy_level(n, h, m, a):
    """Compute the energy level for the nth quantum state in Joules."""
    return (n**2 * np.pi**2 * h**2) / (2 * m * a**2)

# Plot setup
x = np.linspace(0, a, 500)  # Spatial grid
n_levels = 3  # Number of energy levels to display
fig, ax = plt.subplots(figsize=(6, 10))

for n in range(1, n_levels + 1):
    # Calculate wave function, probability density, and energy
    energy = energy_level(n, h, m_e, a) / eV_to_J  # Convert energy to eV
    wf = wave_function(n, x, a) + energy  # Scale wave function and offset by energy
    prob_density =probability_density(n, x, a)  # Probability density

    # Plot wave function
    ax.plot(x * 1e10, wf, color='red', label=f"Wave Function (n={n})")  # Convert x to Å
    # Plot probability density as a filled region
    ax.fill_between(x * 1e10, energy, energy + prob_density, color='blue', alpha=0.3,
                    label=f"|\u03C8|² (n={n})")
    # Plot energy level as dashed line
    ax.hlines(energy, 0, a * 1e10, linestyles='dashed', colors='red')  # Convert x to Å
    # Add energy level label
    ax.text(a * 1.2 * 1e10, energy, f"$n={n}$\n$E_{n}={energy:.3f}$ eV", color='red', fontsize=10)

# Customize plot
ax.set_title("Infinite Square Well with Probability Densities", fontsize=14)
ax.set_xlim(0, a * 1e10)
ax.set_ylim(0, energy_level(n_levels + 1, h, m_e, a) / eV_to_J * 1.2)
ax.set_xlabel(r"$L_1 = 10 \, \mathrm{\AA}$", fontsize=12)  # Fixed escape sequence warning
ax.set_ylabel("Energy (eV)", fontsize=12)
ax.axvline(0, color='gray', linewidth=3)  # Left wall of the well
ax.axvline(a * 1e10, color='gray', linewidth=3)  # Right wall of the well
ax.grid(False)
ax.legend(loc='upper right')

plt.tight_layout()
plt.show()

```

```python
import numpy as np

# Constantes físicas
hbar = 1.054e-34  # Constante reducida de Planck (J·s)
me = 9.109e-31  # Masa del electrón (kg)
a = 10e-10  # Ancho del pozo (10 Å en metros)
n = 1  # Número cuántico para el estado base

# Fórmula para la energía
E1_hbar = (n**2 * hbar**2 * np.pi**2) / (2 * me * a**2)

# Conversión de Joules a eV
eV_to_J = 1.602e-19  # Conversión de 1 eV a Joules
E1_hbar_eV = E1_hbar / eV_to_J

# Resultados
print(f"Energía en Joules: {E1_hbar:.2e} J")
print(f"Energía en electronvoltios: {E1_hbar_eV:.2f} eV")
```
