# Guide

we are goin to evaluate the data of the resistence of graphine vs the voltage determine by the next function 
\begin{align}
R=& R_C + \frac{L}{We\mu \sqrt{[n_0^2 +(C_g(Vg-V_d)/e)^2]}}
\end{align}

### Variables in the equation:

- $R$: Total resistance of the device.
- $R_C$: Contact resistance.
- $L$: Channel (sample) length.
- $W$: Channel (sample) width.
- $e$: Elementary charge of the electron.
- $\mu$: Carrier mobility.
- $n_0$: Intrinsic charge inhomogeneit
- $C_g$: Gate capacitance per unit area.
- $V_g$: Gate voltage.
- $V_d$: Dirac point voltage or threshold voltage.


```python
import numpy as np 
import matplotlib.pyplot as plt
from lmfit import Parameters , Minimizer, Model
from matplotlib import rcParams
from scipy import constants


#load the data 
data=np.loadtxt('data/data_graphene.txt')
# extract the gate voltage and the resistence 
x=data[:,0]
y=data[:,1]

#plot the experiment data 
plt.plot(x,y, 'o', color='darkred')
plt.xlabel(r'$V_{g}$ (Volts)')
plt.ylabel(r'R ($\Omega$')
```

```python
#Function to analize:

def mob(parames, x, datar=None): #X is a positional argument  while the data is actually optional so that the function returns the model calculation if the data is neglected
    rc=parames['rc']
    mu=parames['mu']
    n0=parames['n0']
    epr=parames['epr']
    d=parames['d']
    L=parames['L']
    W=parames['W']
    Vd=parames['Vd']

    C= constants.epsilon_0*epr/d
    n=C*(x-Vd)/constants.e
    R = rc + L / (W * constants.e * mu * np.sqrt((n0*1e16)**2+n**2))
    if datar is None:
        return R
    else:
        return R - datar 
```

```python
pars= Parameters()

# set up the initial parameters values
pars.add('rc', value=500) #initial values consisten with the physical behaviour 

pars.add('n0', value=0.05)
pars.add('mu',value=5, min=0)
pars.add('Vd', value=0.1)
pars.add('epr', value=3.9, vary=False)#the vary= false are constant values
pars.add('d', value=25e-9, vary=False)
pars.add('L',value=5e-6, vary=False)
pars.add('W', value=2e-6, vary=False)
```

```python
#plotting the inital guess
#plot the initial guess
plt.plot(x,mob(pars,x))
plt.plot(x,y,'o')
```

```python
fitter= Minimizer(mob,pars,fcn_args=(x,y))
result=fitter.minimize()
final=y+result.residual
plt.plot(x,y,'o', color='red', label='data', alpha=0.7)
plt.plot(x,final, color='black', label='best fit')
plt.legend()
```

```python
#Load the data 
data = np.loadtxt('data/data_graphene.txt')

# Extract the gate voltage and the resistance 
x = data[:,0]
y = data[:,1]

# Create randomized uncertainties
# Base uncertainty of 3% plus random variation up to an additional 4%
np.random.seed(42)  # For reproducibility
base_uncertainty = 0.03 * y
random_component = np.random.uniform(0, 0.04, size=len(y)) * y
y_err = base_uncertainty + random_component

# Plot the experiment data with error bars
# Reduced marker size to 3 and made the error bars thinner
plt.errorbar(x, y, yerr=y_err, fmt='o', color='darkred', markersize=3, capsize=2, 
             ecolor='black', elinewidth=0.7)

plt.xlabel(r'$V_{g}$ (Volts)')
plt.ylabel(r'Resistance ($\Omega$)')
plt.title('Graphene Resistance vs. Gate Voltage')

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
