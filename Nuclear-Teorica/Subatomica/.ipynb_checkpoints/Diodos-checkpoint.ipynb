# Formato Grafico

```python
#%% FORMATO GRAFICAS
def formato_grafico(ax=None, decimalx=None, decimaly=None, legend_loc='upper right'):
    if ax is None:
        ax = plt.gca()

    fuente = 'sans'
    ax.title.set_size(14)
    ax.xaxis.label.set_size(16)
    ax.yaxis.label.set_size(16)
    ax.title.set_family(fuente)
    ax.yaxis.label.set_family(fuente)
    ax.xaxis.label.set_family(fuente)
    ax.tick_params(axis='both', labelsize=14)

    plt.rcParams['mathtext.fontset'] = 'cm'
    plt.rcParams['mathtext.rm'] = 'sans'
    plt.rcParams['mathtext.default'] = 'rm'

    ax.xaxis.set_major_locator(ticker.AutoLocator())
    ax.xaxis.set_minor_locator(ticker.AutoMinorLocator())
    ax.yaxis.set_major_locator(ticker.AutoLocator())
    ax.yaxis.set_minor_locator(ticker.AutoMinorLocator())

    ax.tick_params(axis='x', direction='in', which='both', bottom=True, top=True)
    ax.tick_params(axis='y', direction='in', which='both', left=True, right=True)
    ax.tick_params(axis='x', which='minor', direction='in', bottom=True, top=True)
    ax.tick_params(axis='y', which='minor', direction='in', left=True, right=True)

    # Configurar la notaci√≥n cient√≠fica
    if decimalx is not None:
        ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{x:.{decimalx}f}' if x != 0 else '0'))

    if decimaly is not None:
        ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{x:.{decimaly}f}' if x != 0 else '0'))

    if ax.get_legend() is not None:
        ax.legend(fontsize=16)

    font = font_manager.FontProperties(family='sans', style='normal', size=16)

    handles, labels = ax.get_legend_handles_labels()
    if handles and labels:
        ax.legend(loc=legend_loc, handles=handles, labels=labels, edgecolor='1', framealpha=0.5, prop=font)
```

#Data & Uncertanties

```python
def load_diode_data_with_uncertainty(file_path):
    """
    Carga datos de voltaje en resistencia y voltaje de diodo.
    Calcula la corriente usando la Ley de Ohm con propagaci√≥n de incertidumbre.

    Par√°metros:
    - file_path: ruta al archivo con dos columnas (V_R, V_d)

    Retorna:
    - V_diodo: arreglo de ufloat (con incertidumbre fija de 0.01 V)
    - I_diodo: corriente calculada como V_R / R (con incertidumbres propagadas)
    """
    V_res_list = []
    V_d_list = []

    with open(file_path, 'r', encoding='latin1') as f:
        for line in f:
            parts = line.strip().replace(',', ' ').split()
            if len(parts) != 2:
                continue  # saltar l√≠neas mal formateadas
            try:
                V_res = float(parts[0])
                V_d = float(parts[1])
                V_res_list.append(V_res)
                V_d_list.append(V_d)
            except ValueError:
                continue  # saltar l√≠neas con datos no num√©ricos

    V_res = np.array(V_res_list)
    V_d = np.array(V_d_list)

    # Aplicar incertidumbre:
    V_res_unc = unp.uarray(V_res, 0.10 * V_res)  # 10% para V_resistencia
    V_d_unc = unp.uarray(V_d, 0.01)              # 0.01 V para V_diodo

    # Resistencia con 5% de tolerancia
    R = ufloat(1500, 1500 * 0.05)  # 1500 ¬± 75 ohmios

    # Corriente en el diodo (con propagaci√≥n de incertidumbre)
    I_d = V_res_unc / R  # Ley de Ohm

    return V_d_unc, I_d
```

# ODR function

```python
#def rect(x, b0, b1):
    #'''
    #Descripci√≥n
    #-----------
    #C√°lculo de una funci√≥n lineal a partir de unos par√°metros dados.

    #Par√°metros
   # ----------
    #x (float or np.array): Variable dependiente
   # b0 (float): Corte con el eje y
   # b1 (float): Pendiente


    #Retorna
    #-------
    #Arreglo (o valor individual) de la funci√≥n evaluada en x
    #'''
   # return b0 + b1*(x)
def rect(p, x):
    return p[1] * x + p[0]
def ODR_custom_fit(x_with_unc, y_with_unc, xmin, xmax, b0,  modelo=rect, fix_intercept = False):
    # Handle x
    if isinstance(x_with_unc[0], UFloat):
        x_nom = unp.nominal_values(x_with_unc)
        x_err = unp.std_devs(x_with_unc)
    else:
        x_nom = np.array(x_with_unc, dtype=float)
        x_err = np.full_like(x_nom, 1e-10)

    # Handle y
    if isinstance(y_with_unc[0], UFloat):
        y_nom = unp.nominal_values(y_with_unc)
        y_err = unp.std_devs(y_with_unc)
    else:
        y_nom = np.array(y_with_unc, dtype=float)
        y_err = np.zeros_like(y_nom)

    # Subset
    mask = (x_nom >= xmin) & (x_nom <= xmax)
    x_fit = x_nom[mask]
    x_fit_err = x_err[mask]
    y_fit = y_nom[mask]
    y_fit_err = y_err[mask]

    # --------- Mejores guesses para ODR ----------
    if len(x_fit) >= 2:
        # Estimar pendiente con primer y √∫ltimo punto del rango
        b1_guess = (y_fit[1] - y_fit[0]) / (x_fit[1] - x_fit[0])
    else:
        b1_guess = 1.0  # valor por defecto si hay muy pocos puntos

    b0_guess = b0
    beta0 = [b0_guess, b1_guess]

    # ---------------------------------------------
    # Build model and data for ODR
    #def odr_wrapper(f):
      #  def wrapped(B, x):
          #  return f(x, *B)
        #return wrapped

    Rmodel = ODRModel(modelo)#ODRModel(odr_wrapper(modelo))
    data = RealData(x_fit, y_fit, sx=x_fit_err, sy=y_fit_err)
    if fix_intercept:
      odr = ODR(data, Rmodel, beta0=beta0, ifixb=[0, 1])
    else:
      odr = ODR(data, Rmodel, beta0=beta0)

    return odr, data
```

```python
def fit_report(data, odr):
    """
    Genera gr√°fica del ajuste + fit_report en formato Markdown extendido.
    Compatible con modelos de un pico y doble pico.

    Parameters:
    - result: objeto de resultado de lmfit
    - x, y: datos experimentales (sin incertidumbre)
    - x_ajuste: eje x del modelo ajustado
    - y_ajuste: modelo completo (recta + gauss)
    - recta_ajuste: modelo de fondo lineal
    - titulo: t√≠tulo del gr√°fico y nombre del archivo
    - unidades: 'canal' o 'keV'
    - ylim: l√≠mites opcionales para eje y
    - save_path: ruta donde guardar la imagen
    - show_plot: si se debe mostrar el gr√°fico en pantalla

    Returns:
    - Markdown fit report como objeto renderizable en Jupyter
    """
    result = odr.run()
    Beta =[]
    Beta_err =[]
    Beta = list(result.beta)
    Beta_err = list(result.sd_beta)
    # Recuperar par√°metros ajustados con incertidumbre
    b0 = ufloat(Beta[0], Beta_err[0])
    b1 = ufloat(Beta[1], Beta_err[1])

    # ----- CALCULAR datos estadisticos -----
    # when usin iwork[i], if n is the number of parameters i = 4 + 2n is the number of data points
    # i = 5 + 2n is the number of parameters and i = 6 +2n is the number of degrees of freedom
    i = 6 +2*len(Beta)
    df = result.iwork[i]
    Chi2red = result.res_var
    Chi2 = Chi2red*df
    Q = chi2.sf(Chi2, df).astype(np.float64)

    return b0 , b1 , Chi2red, Chi2, Q,
```

### ‚öôÔ∏è Ecuaci√≥n de Shockley no ideal ‚Äì Linealizaci√≥n

La **ecuaci√≥n de Shockley no ideal** para un diodo se expresa como:

$$
I(V) = I_0 \left( e^{\frac{qV}{\eta k_B T}} - 1 \right)
$$

donde:  
- $( I(V) )$: corriente a trav√©s del diodo  
- $( I_0 )$: corriente de saturaci√≥n inversa  
- $( q )$: carga del electr√≥n $( 1.602 \times 10^{-19} \, \text{C} )$  
- $( V )$: voltaje en el diodo  
- $( k_B $): constante de Boltzmann $( 1.381 \times 10^{-23} \, \text{J/K} )$  
- $( T )$: temperatura absoluta (Kelvin)  
- $( \eta )$: **factor de idealidad** del diodo $( 1 \leq \eta \leq 2 )$  

---

#### üîÑ Suposici√≥n para simplificaci√≥n:

En la regi√≥n **directa y exponencial del diodo**:

$$
V \gg \frac{\eta k_B T}{q} \quad \Rightarrow \quad e^{\frac{qV}{\eta k_B T}} \gg 1
$$

por lo tanto:

$$
I(V) \approx I_0 \, e^{\frac{qV}{\eta k_B T}}
$$

---

#### üßÆ Tomando logaritmo natural:

$$
\ln I \approx \ln \left( I_0 \, e^{\frac{qV}{\eta k_B T}} \right)
\quad \Rightarrow \quad \ln I \approx \ln I_0 + \frac{qV}{\eta k_B T}
$$

---

### ‚úÖ Ecuaci√≥n lineal final:

$$
\ln I = \ln I_0 + \frac{q}{\eta k_B T} V
$$

Esta es una ecuaci√≥n de la forma:

$$
y = a + b x
$$

con:  
- $( y = \ln I )$  
- $ x = V $  
- pendiente $ b = \frac{q}{\eta k_B T} $
- ordenada al origen $ a = \ln I_0 $


```python
file_path='/content/drive/Shareddrives/Electronica Anal/Electronica Analogica/datos/datos resistencia dinamica diodos/V_resitencia_vs_V_diodoIn4004.dat'
# Cargar y ajustar
V_d, I_d = load_diode_data_with_uncertainty(file_path)
x_min = 0
x_max = 0.8
q = 1.602e-19      # C
kB = 1.381e-23     # J/K
T = 290            # K (ajusta si usas otra temperatura)
def modelo_lnI(x, ln_I0, eta):

  return ln_I0 + (q / (eta * kB * T)) * x
# Transformaci√≥n: ln(I)
ln_I_d = unp.log(I_d)
ln_I_d_nom = unp.nominal_values(ln_I_d)
ln_I_d_err = unp.std_devs(ln_I_d)
V_d_nom = unp.nominal_values(V_d)
V_d_err = unp.std_devs(V_d)
# Filter out non-positive values for logarithm
valid_indices = ln_I_d_nom > 0
ln_I_d_filtered = ln_I_d[valid_indices]
V_d_filtered = V_d[valid_indices]
#

# I0 fijo
I0_fixed = 5.0e-6
ln_I0 = np.log(I0_fixed)


#Modelo 1: fijar I_0
Odr_function1, real_data1 = ODR_custom_fit(V_d_filtered, ln_I_d_filtered,  xmin= x_min, xmax= x_max, b0= ln_I0, modelo= rect, fix_intercept= True )
ln_I0_1, b1, Chi2red1, Chi21, Q1 = fit_report(real_data1, Odr_function1)
eta1  =  q/(b1*kB *T)

#Modelos 2: I0 libre
Odr_function2, real_data2 = ODR_custom_fit(V_d_filtered, ln_I_d_filtered, xmin=x_min, xmax=x_max, b0= 0.5, modelo= rect, fix_intercept= False )
lnI0_2, b2, Chi2red2, Chi22, Q2 = fit_report(real_data2, Odr_function2)
eta2  = q/(b2 *kB *T)
I0_2 = unp.exp(lnI0_2)
# Generar curva de ajuste
x_line = np.linspace(x_min, x_max, 200)
fit1 = modelo_lnI(x_line, ln_I0, eta1.nominal_value)
fit2 = modelo_lnI(x_line, lnI0_2.nominal_value, eta2.nominal_value)

# Gr√°fico
plt.figure(figsize=(8, 6))
plt.errorbar(V_d_nom, ln_I_d_nom, xerr=V_d_err, yerr=ln_I_d_err, fmt='o', capsize=4, label='Datos ln(I)')
plt.plot(x_line, fit1, 'r--', label=f'Ajuste: $I_0$ fijo, Œ∑ = {eta1:.2uP}')
plt.plot(x_line, fit2, 'g-', label=f'Ajuste: $I_0$ libre = {I0_2:.2uP} A, Œ∑ = {eta2:.2uP}')
plt.xlabel("Voltaje en el diodo (V)")
plt.ylabel("ln(Corriente en el diodo) [ln(A)]")
plt.grid(True)
plt.legend()
plt.tight_layout()

# Guardar
output_dir = '/content/drive/MyDrive/Electronica Analogica/codigo/Diodos'
os.makedirs(output_dir, exist_ok=True)
plt.savefig(os.path.join(output_dir, 'ajuste_lnI_vs_V_shockley.pdf'), dpi=300)
plt.show()

print("Modelo 1: I0 fijo")
print(f"Œ∑ = {eta1:.3uP}")
print(f"I0 = {ln_I0_1:.3uP}")
print(f"chi2 reducido = {Chi2red1:.3f}")
print(f"Chi2 total = {Chi21:.3f}")
print(f"Q = {Q1:.3f}")

print("\nModelo 2: I0 libre")
print(f"Œ∑ = {eta2:.3uP}")
print(f"I0 = {I0_2:.3uP} A")
print(f"ln(I0) = {lnI0_2:.3uP}")
print(f"chi2 reducido = {Chi2red2:.3f}")
print(f"Chi2 total = {Chi22:.3f}")
print(f"Q = {Q2:.3f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import odr
import uncertainties as unc
from uncertainties import unumpy
file_path='/content/drive/Shareddrives/Electronica Anal/Electronica Analogica/datos/datos resistencia dinamica diodos/V_resitencia_vs_V_diodoIn4004.dat'

# Cargar y ajustar
V_d, I_d = load_diode_data_with_uncertainty(file_path)

# Extract nominal values and standard deviations for ODR
V_d_nom = unumpy.nominal_values(V_d)
V_d_std = unumpy.std_devs(V_d)
I_d_nom = unumpy.nominal_values(I_d)
I_d_std = unumpy.std_devs(I_d)

# Apply natural logarithm to current data
# Filter out non-positive values for logarithm
valid_indices = I_d_nom > 0
ln_I_d_nom = np.log(I_d_nom[valid_indices])
V_d_nom_filtered = V_d_nom[valid_indices]
V_d_std_filtered = V_d_std[valid_indices]

# Calculate uncertainty in ln(I) using error propagation
ln_I_d_std = I_d_std[valid_indices] / I_d_nom[valid_indices]

# Define linear function for ODR
def linear_func(p, x):
    return p[0] * x + p[1]

# Create ODR model
linear_model = odr.Model(linear_func)
data = odr.RealData(V_d_nom_filtered, ln_I_d_nom, sx=V_d_std_filtered, sy=ln_I_d_std)
odr_obj = odr.ODR(data, linear_model, beta0=[1.0, -12.0], ifixb=[1, 0])
output = odr_obj.run()

# Get parameters with uncertainties
slope = unc.ufloat(output.beta[0], output.sd_beta[0])
intercept = unc.ufloat(output.beta[1], output.sd_beta[1])

# Calculate related diode parameters
kT_q = 1/slope  # Thermal voltage in volts
Is =u.wrap(np.exp)(intercept)  # Saturation current


# Create figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plot ln(I) vs V (linear model)
ax1.errorbar(V_d_nom_filtered, ln_I_d_nom, xerr=V_d_std_filtered, yerr=ln_I_d_std,
            fmt='o', label='Data', markersize=4, capsize=3)

# Plot the fitted line
x_fit = np.linspace(min(V_d_nom_filtered), max(V_d_nom_filtered), 100)
y_fit = linear_func(output.beta, x_fit)
ax1.plot(x_fit, y_fit, 'r-', label=f'Fit: ln(I) = ({slope:.3f})V + ({intercept:.3f})')

ax1.set_xlabel('Voltage [V]')
ax1.set_ylabel('ln(Current) [ln(A)]')
ax1.set_title('Linear Fit of ln(I) vs V')
ax1.grid(True)
ax1.legend()

# Plot I vs V (original data with exponential model)
ax2.errorbar(V_d_nom, I_d_nom, xerr=V_d_std, yerr=I_d_std,
            fmt='o', label='Data', markersize=4, capsize=3)

# Plot the fitted exponential model
y_exp_fit = np.exp(output.beta[1]) * np.exp(output.beta[0] * x_fit)
ax2.plot(x_fit, y_exp_fit, 'r-', label=f'Fit: I = {Is:.3e} * exp(V/{kT_q:.4f})')

ax2.set_xlabel('Voltage [V]')
ax2.set_ylabel('Current [A]')
ax2.set_title('Exponential Fit of I vs V')
ax2.grid(True)
ax2.legend()

plt.tight_layout()

# Print results
print("ODR Linear Fit Results:")
print(f"Slope = {slope}")
print(f"Intercept = {intercept}")
print("\nDiode Parameters:")
print(f"Thermal voltage (kT/q) = {kT_q} V")
print(f"Saturation current (Is) = {Is} A")
print(f"Ideality factor (assuming T = 300K) = {kT_q/(0.026)} ")

plt.show()
```
